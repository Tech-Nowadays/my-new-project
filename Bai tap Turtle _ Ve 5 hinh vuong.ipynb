```python

import turtle
window = turtle.Screen()
turtle.speed(5)

#draw an outline of a square
turtle.penup()
turtle.goto(2,40)
turtle.pendown()

turtle.forward(150)
turtle.left(90)
turtle.forward(150)
turtle.left(90)
turtle.forward(150)
turtle.left(90)
turtle.forward(150)
turtle.left(90)
turtle.penup()
#draw the outline of a square using a for loop
turtle.goto(-70,-10)
turtle.pendown()
for i in range (4) :
    turtle.forward(150)
    turtle.left(90)
turtle.penup()

#draw the 3rd square
turtle.goto(-140,40)
turtle.pendown()
for i in range (4) :
    turtle.forward(150)
    turtle.left(90)
turtle.penup()

#draw the 4th square
turtle.goto(-130,150)
turtle.pendown()
for i in range (4) :
    turtle.forward(150)
    turtle.left(90)
turtle.penup()

#draw the 5th square
turtle.goto(1,150)
turtle.pendown()
for i in range (4) :
    turtle.forward(150)
    turtle.left(90)
turtle.penup()

#add an dot in the center
turtle.goto(3,120)
turtle.pendown()
turtle.fillcolor('black')
turtle.begin_fill()
turtle.circle(20)
turtle.end_fill()
turtle.hideturtle()

turtle.done()
```
